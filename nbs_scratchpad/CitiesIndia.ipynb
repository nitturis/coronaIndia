```python
import json

import pandas as pd
```

```python
web_df = pd.read_html(
    "https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population"
)
```

```python
city_names = web_df[0]["City"].tolist()
```

```python
def clean_numbers(city_name: str):
    return "".join([i for i in city_name if i.isalpha()])


clean_city_names = [clean_numbers(city) for city in city_names]
print(json.dumps(clean_city_names, indent=2))
```
